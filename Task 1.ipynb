```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv(r'C:\Users\chira\Downloads\e.csv')
print("Dataset is uploaded, Go ahead")
```

```python
india = df[df['Country Name'] == 'India']
years = india.columns[4:] 
population = india.iloc[0, 4:].values

plt.figure(figsize=(12,6))
plt.bar(years, population, color='skyblue')
plt.title("India's Increase in Population (1960–2024)", fontsize=16)
plt.xlabel("Year")
plt.ylabel("Population (in Billions)")
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

```python
df_sorted = df.sort_values(by='2024', ascending=False).head(10)

plt.figure(figsize=(10,6))
plt.bar(df_sorted['Country Name'], df_sorted['2024'], color='orange')
plt.title("Population Distribution According to Different Categories", fontsize=16)
plt.xlabel("Country")
plt.ylabel("Population (in Billions)")
plt.xticks(rotation=45)
plt.show()
```
